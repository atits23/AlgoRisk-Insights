```python
hindustan=pd.read_csv('hindustan.csv')
```

```python
hindustan['Moving_Average']=hindustan['Close'].rolling(window=20).mean()
```

```python
hindustan.plot.line(x='Date',y=['Close','Moving_Average'])
```

#
Code For 10-Days SMA

```python
moving_avg=[]
for i in range(243):
    if i<9:
        moving_avg.append(float('NaN'))
    else:
        average=sum(hindustan['Close'].iloc[i-9:i+1])/10
        moving_avg.append(average)

hindustan['SMA_10']=pd.DataFrame(moving_avg)
```

```python
hindustan
```

```python
hindustan.plot.line(x='Date', y=['Close', 'Moving_Average', 'SMA_10'], figsize=(10, 6))
```
