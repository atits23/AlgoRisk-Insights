```python
import numpy as np
import pandas as pd
import yfinance as yf
import warnings
import datetime
warnings.filterwarnings('ignore')
import scipy.stats as stats
```

```python
data=pd.read_csv('all_stocks_5yr.csv')
```

```python
columns=np.array(data['Name'].unique())
```

```python
data
```

```python
data = data.pivot(index='date', columns='Name', values='close')
```

```python
data
```

```python
returns=data.pct_change()
```

```python
confidence_level = 0.95
mean = returns.mean()
sigma = returns.std()
```

```python
mean
```

```python
mean={
    'Mean':mean
}
mean=pd.DataFrame(mean)
std={
    'Sigma':sigma
}
sigma=pd.DataFrame(std)
```

```python
sigma
```

```python
VaR=[]
CoVaR=[]
```

```python
for i in range(len(mean)):
    Z_alpha = stats.norm.ppf(1 - confidence_level)
    V = mean['Mean'].iloc[i] + Z_alpha *sigma['Sigma'].iloc[i]
    VaR.append(V*100)
    phi_Z_alpha = stats.norm.pdf(Z_alpha)
    Co = mean['Mean'].iloc[i] + (phi_Z_alpha / (1 - confidence_level)) *sigma['Sigma'].iloc[i]
    CoVaR.append(Co*100)
```

```python
df={
    'Ticker':columns,
    'VaR':VaR,
    'CoVaR':CoVaR
}
df=pd.DataFrame(df)
```
