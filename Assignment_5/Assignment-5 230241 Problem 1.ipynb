```python
import numpy as np
import pandas as pd
import yfinance as yf
import warnings
import datetime
warnings.filterwarnings('ignore')
from scipy.stats import jarque_bera
```

```python
data=pd.read_csv('all_stocks_5yr.csv')
```

```python
data
```

```python
columns=np.array(data['Name'].unique())
```

```python
data = data.pivot(index='date', columns='Name', values='close')
```

```python
returns=data.pct_change().dropna()
```

```python
jb_statistics = []
jb_p_values = []
```

```python
for col in columns:
    jb_stat, jb_p = jarque_bera(returns[col])
    jb_statistics.append(jb_stat)
    jb_p_values.append(jb_p)
```

```python
df={
    'ticker':columns,
     'jb_stat':jb_statistics,
     'jb_p_values':jb_p_values

}
df=pd.DataFrame(df)
```

```python
df
```

```python
df['check'] = np.where(df['jb_p_values'] > 0.05, 'Normal', 'N-Normal')
df
```
